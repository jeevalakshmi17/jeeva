```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
cData = pd.read_csv("/content/stocks.csv")
```

```python
#dropping


# Dropping rows with NaN values in the 'Date' column
cData = cData.drop(columns=['Date'])
cData = cData.drop(columns=['Ticker'])
# Check the data info after dropping
cData.info()
```

```python
# lets build our linear model
# independant variables
X = cData.drop(['Volume','Open'], axis=1)
# the dependent variable
y = cData[['Volume']]
```

```python
X.head()
```

```python
X.shape
```

```python
# Split X and y into training and test set in 70:30 ratio

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model

poly = PolynomialFeatures(degree=2, interaction_only=True)
X_train2 = poly.fit_transform(X_train)
X_test2 = poly.fit_transform(X_test)

poly_clf = linear_model.LinearRegression()

poly_clf.fit(X_train2, y_train)

y_pred = poly_clf.predict(X_test2)

#print(y_pred)

#In sample (training) R^2 will always improve with the number of variables!
print(poly_clf.score(X_train2, y_train))
```

```python
#Out off sample (testing) R^2 is our measure of sucess and does improve
print(poly_clf.score(X_test2, y_test))
```

```python
# but this improves as the cost of 7 extra variables!
print(X_train.shape)
print(X_train2.shape)
```
